```python
import pandas as pd
import random

# Pisahkan data berdasarkan jenis kelamin
pria = data[data['JENISKELAMIN'] == 'PRIA']  # Dataframe berisi mahasiswa PRIA
wanita = data[data['JENISKELAMIN'] == 'WANITA']  # Dataframe berisi mahasiswa wanita

# Randomisasi data pria dan wanita
random.shuffle(pria)
random.shuffle(wanita)

# Buat 9 kelompok kosong
kelompok = [[] for _ in range(9)]

# Tambahkan maksimal 3 pria per kelompok
for i in range(len(pria)):
    kelompok[i % 9].append(pria[i])

# Tambahkan minimal 2 wanita per kelompok
for i in range(len(wanita)):
    kelompok[i % 9].append(wanita[i])

# Tampilkan hasil
for i, k in enumerate(kelompok):
    print(f"Kelompok {i+1}:")
    for member in k:
        print(member)
```
